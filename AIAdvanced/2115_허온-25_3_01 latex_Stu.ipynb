### ◈ 실수 a의 절댓값의 정의<br>
$\left\vert a \right\vert =
\begin{Bmatrix}
a & (a \ge 0)\\
-a & (a<0)
\end{Bmatrix}$<br>
### ◈ 절댓값의 성질 : a, b가 실수일 때<br>
$\left\vert a \right\vert \ge 0, \left\vert -a \right\vert = \left\vert a \right\vert$<br>
$\left\vert a \right\vert^2 = a^2, \left\vert a \right\vert \left\vert b \right\vert = \left\vert ab \right\vert$<br>
$\frac{\left\vert a \right\vert}{\left\vert b \right\vert} = \left\vert \frac{a}{b} \right\vert(b\neq0)$

### 방정식 $x^3 = 1$의 허근의 성질
◈ $x^3 - 1 = (x-1) (x^2 + x + 1)$ \\
◈ 방정식 $x^3 = 1$의 한 허근을 $\omega$라 하면 \\
$\quad$    1) $\omega^3 = \omega^2 + \omega + 1$ \\
$\quad$    2) $\omega$의 켤레 복소수 $\bar{\omega}$도 해이다. \\
$\quad$    3) $\omega + \bar{\omega} = -1, \omega \bar{\omega} = 1$ \\
$\quad$    4)  $\omega^2 = -\omega -1 = \bar{\omega} = \frac{1}{\omega}$
