```python
loppy_face = img2[100:500, 140:580] # 400, 440

def resize(img):
  ylength = len(img) // 2
  xlength = len(img[0]) // 2
  resized_img = np.zeros((ylength, xlength, 3), dtype=np.uint8)

  for i in range(ylength):
    for j in range(xlength):
        value1 = img[2 * i, 2 * j] / 4
        value2 = img[2 * i + 1, 2 * j] / 4
        value3 = img[2 * i, 2 * j + 1] / 4
        value4 = img[2 * i + 1, 2 * j + 1] / 4

        total = value1 + value2 + value3 + value4

        resized_img[i][j] = total

  return resized_img

double_reseized_loppy_head = resize(resize(resize(loppy_face)))
cv2_imshow(double_reseized_loppy_head)
```
