```python
import numpy as np
from scipy.io.wavfile import write
from IPython.display import Audio

def get_sin_sound(frequency):
  # Define parameters for the sine wave
  sampling_rate = 44100  # Hz
  duration = 2  # seconds
  amplitude = 0.5  # range from 0 to 1

  # Generate time values
  t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
  # xì¢Œí‘œ(ì‹œê°„ ë°°ì—´)

  # Generate the sine wave
  # 2pië¥¼ ê³±í•œê±´ sinì˜ ì£¼ê¸°ë¥¼ 1ë¡œ ë§ì¶”ê³  frenquencyë§Œí¼ ì¡°ì ˆí•˜ê¸° ìœ„í•¨
  sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)

  print("x ì¢Œí‘œ ëª¨ì„",t)
  print("y ì¢Œí‘œ ëª¨ì„", sine_wave)


  # Scale the sine wave to 16-bit integer format


  scaled_sine_wave = np.int16(sine_wave * 32767)
  #  -32767~32767ì˜ ì •ìˆ˜ìë£Œë¥¼ ê°€ì§€ëŠ” ë°°ì—´í™”


  # Save the sine wave as a WAV file
  output_file = 'sine_wave.wav'
  write(output_file, sampling_rate, scaled_sine_wave)

  # Play the generated audio
  display(Audio(output_file))

# ë„
get_sin_sound(262)

# ë¯¸
get_sin_sound(330)

# ì†”
get_sin_sound(392)
```

```python
import numpy as np
from scipy.io.wavfile import write
from IPython.display import Audio

# Define parameters for the sine wave
sampling_rate = 44100  # Hz
duration = 2  # seconds
frequency = 880  # Hz (A4 note)
amplitude = 0.5  # range from 0 to 1

# Generate time values
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
# xì¢Œí‘œ(ì‹œê°„ ë°°ì—´)

# Generate the sine wave
# 2pië¥¼ ê³±í•œê±´ sinì˜ ì£¼ê¸°ë¥¼ 1ë¡œ ë§ì¶”ê³  frenquencyë§Œí¼ ì¡°ì ˆí•˜ê¸° ìœ„í•¨
sine_wave1 = amplitude * np.sin(2 * np.pi * 262 * t)
sine_wave2 = amplitude * np.sin(2 * np.pi * 330 * t)
sine_wave3 = amplitude * np.sin(2 * np.pi * 392 * t)

# ë‘ ìˆ˜ë¥¼ ë”í•˜ê¸°
combine_sine_wave = sine_wave1 + sine_wave2 + sine_wave3

# ìµœëŒ“ê°’ì„ 1ë¡œ ë§ì¶°ì¤˜ì„œ amplitudeê°€ íš¨ê³¼ìˆë„ë¡ ìˆ˜ì •
combine_sine_wave = combine_sine_wave / combine_sine_wave.max()

scaled_sine_wave = np.int16(combine_sine_wave * 32767)
#  -32767~32767ì˜ ì •ìˆ˜ìë£Œë¥¼ ê°€ì§€ëŠ” ë°°ì—´í™”


# Save the sine wave as a WAV file
output_file = 'sine_wave.wav'
write(output_file, sampling_rate, scaled_sine_wave)

# Play the generated audio
Audio(output_file)
```
