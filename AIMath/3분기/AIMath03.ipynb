```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio

# Define the path to your MP3 file
mp3_file_path = '/content/무료노래.mp3' # Replace with your actual file path

# Load the MP3 file
try:
    y, sr = librosa.load(mp3_file_path)

    # Print the shape of the NumPy array
    print(f"Shape of the audio array: {y.shape}")
    print(f"Sampling rate: {sr}")

    print(y[1*(1960686//3):2*(1960686//3)])


    # 파형 그리기(pyplotlib, 1초정도)
    plt.plot(y)


    # 볼륨 조절하기
    mini_y = y / 2
    plt.plot(mini_y, color='r')

    # 속도 조절하기
    fast_y = []
    for i in range(0, len(y), 2):
      new_value = y[i] + y[i + 1] / 2
      fast_y.append(new_value)

    plt.plot(fast_y, color='g')

    plt.show()

    print("🎵 원본 소리")
    display(Audio(y, rate=sr))

    print("🎵 작은 소리")
    display(Audio(mini_y, rate=sr))

    print("🎵 빠른 소리")
    display(Audio(fast_y, rate=sr))


except Exception as e:
    print(f"Error loading the MP3 file: {e}")
```

```python
import numpy as np
from scipy.io.wavfile import write
from IPython.display import Audio

def get_sin_sound(frequency):
  # Define parameters for the sine wave
  sampling_rate = 44100  # Hz
  duration = 2  # seconds
  amplitude = 0.5  # range from 0 to 1

  # Generate time values
  t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
  # x좌표(시간 배열)

  # Generate the sine wave
  # 2pi를 곱한건 sin의 주기를 1로 맞추고 frenquency만큼 조절하기 위함
  sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)

  print("x 좌표 모임",t)
  print("y 좌표 모임", sine_wave)


  # Scale the sine wave to 16-bit integer format


  scaled_sine_wave = np.int16(sine_wave * 32767)
  #  -32767~32767의 정수자료를 가지는 배열화


  # Save the sine wave as a WAV file
  output_file = 'sine_wave.wav'
  write(output_file, sampling_rate, scaled_sine_wave)

  # Play the generated audio
  display(Audio(output_file))

# 도
get_sin_sound(262)

# 미
get_sin_sound(330)

# 솔
get_sin_sound(392)
```

### 화음 넣기

```python
import numpy as np
from scipy.io.wavfile import write
from IPython.display import Audio

# Define parameters for the sine wave
sampling_rate = 44100  # Hz
duration = 2  # seconds
frequency = 880  # Hz (A4 note)
amplitude = 0.5  # range from 0 to 1

# Generate time values
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
# x좌표(시간 배열)

# Generate the sine wave
# 2pi를 곱한건 sin의 주기를 1로 맞추고 frenquency만큼 조절하기 위함
sine_wave1 = amplitude * np.sin(2 * np.pi * 262 * t)
sine_wave2 = amplitude * np.sin(2 * np.pi * 330 * t)
sine_wave3 = amplitude * np.sin(2 * np.pi * 392 * t)

# 두 수를 더하기
combine_sine_wave = sine_wave1 + sine_wave2 + sine_wave3

# 최댓값을 1로 맞춰줘서 amplitude가 효과있도록 수정
combine_sine_wave = combine_sine_wave / combine_sine_wave.max()

scaled_sine_wave = np.int16(combine_sine_wave * 32767)
#  -32767~32767의 정수자료를 가지는 배열화


# Save the sine wave as a WAV file
output_file = 'sine_wave.wav'
write(output_file, sampling_rate, scaled_sine_wave)

# Play the generated audio
Audio(output_file)
```
